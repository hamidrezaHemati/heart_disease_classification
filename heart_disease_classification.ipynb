# A. training the model using decision trees
### 1. extract the cardio column as target vector
### 2. making test and train sets from dataset
### 3. training and testing the model
### 4. tuning the decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

### 1.

```python
y = df.cardio
X = df.drop(['cardio'] , axis =1)
```

### 2. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
print('Training dataset size:', X_train.shape)
print('test dataset size:', X_test.shape)
```

### 3. 

```python
classifier = DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train, y_train)
```

```python
predic_train = classifier.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
classifier_accuracy = accuracy_score(y_train, predic_train)   # accuracy score
cm_classifier = confusion_matrix(y_train, predic_train) # Confusion matrix 

print("Accuracy:",classifier_accuracy)

print('\n Confusion matrix \n \n')
print(classification_report(y_train, predic_train))
```

```python
prediction = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
classifier_accuracy = accuracy_score(y_test, prediction)   # accuracy score
cm_classifier = confusion_matrix(y_test, prediction) # Confusion matrix 

print("Accuracy:",classifier_accuracy)

print('\n Confusion matrix \n \n')
print(classification_report(y_test, prediction ))
```

## Modifying the algorithm for better result and accuracy
as we saw above the accuracy of the train set is 97% and the accuracy of test set is 63%.
this means our model is overfited to the train set. in order to tune the model and get better result we should increase the bias and decrease the varians of our model
to do this we shoud set a maximum depth restriction for the model.

```python
print(classifier.get_depth())
```

```python
from matplotlib.legend_handler import HandlerLine2D
train_acc1 = []
val_acc1 = []

for max_d in range(1,21):
  model = DecisionTreeClassifier(max_depth=max_d, random_state=42, criterion='entropy')
  model.fit(X_train, y_train)
  train_acc1.append(model.score(X_train, y_train))
  val_acc1.append(model.score(X_test,y_test))

line1, = plt.plot([*range(1,21)], train_acc1, 'b', label='Train accuracy')
line2, = plt.plot([*range(1,21)], val_acc1, 'r', label='test accuracy')

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.title('Train vs test Accuracy : Max_depth')
plt.ylabel('Accuracy')
plt.xlabel('Max_depth')
plt.show()

train_acc1.clear()
val_acc1.clear()
```
